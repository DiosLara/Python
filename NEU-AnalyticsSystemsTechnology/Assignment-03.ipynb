```python
### GRADED
# The following piece of code creates a DataFrame 
# (a) Determine the shape of the dataframe
# (b) Print out all the columns of the dataframe
# (c) Print the 3rd element of the dataframe
# (d) Find the average of 'petal_width' where species is 'virginica'
# (e) Find the maximum of 'sepal_width' where species is 'setosa'
# (f): What is the average value of sepal_length
# (g): What is the maximum value of sepal_width
# (h): What is the minimum value of petal_width
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df.head(10) #prints the first 10 rows


# (a) 
print('Shape is:',df.shape)
print('---------')

# (b) 
print('Columns are:',df.columns)
print('---------')

# (c) 
print('3RD ELEMENT OF THE DATAFRAME IS 2ND INDEX:')
print(df.iloc[2])
print('---------')

# (d) 
dfspieces = df.groupby(['species'])
m= dfspieces.get_group('virginica')['petal_width'].mean()
print('Average of petal_width where species is virginica:',m)
print('---------')

# (e) 
dfspieces = df.groupby(['species'])
n= dfspieces.get_group('setosa')['sepal_width'].max()
print('Maximum of sepal_width where species is setosa:',n)
print('---------')

# (f): 
f= round(df.sepal_length.mean(),2)
print('Average value of sepal_length',f)
print('---------')

# (g): 
g= round(df.sepal_width.max(),2)
print('Max value of sepal_width',g)
print('---------')

# (h): 
h= round(df.petal_width.max(),2)
print('Min value of petal_width',h)
```
